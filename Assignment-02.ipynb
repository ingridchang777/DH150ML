```python
import sklearn.linear_model
model = sklearn.linear_model.LinearRegression()
```

```python
import sklearn.model_selection
```

```python
import numpy as np
```

```python
feat2 = np.array(feat2).reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    feat2, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test = x_test.reshape(-1,1)
x_train = x_train.reshape(-1,1)
```

```python
model.fit(x_train,y_train)
```

```python
x_sample = x_test
y_pred = model.predict(x_test)
for i in range(len(y_test)):
    print(y_test[i], y_pred[i])
```

```python
plt.scatter(x_test,y_test)

x_new = np.linspace(-.1,.15,10000)
x_new = x_new.reshape(-1,1)
y_pred = model.predict(x_new)
plt.plot(x_new, y_pred)

plt.show()
```

```python
print(model.coef_, model.intercept_)
```

```python
plt.plot(x_train,y_train,'bo')
plt.plot(x_test,y_test,'ko')

x_new = np.linspace(-.1,.15,10000)
x_new = x_new.reshape(-1,1)

# the predicted y values are now from a model equation, 
# not from results of calling the predict function
y_pred = model.intercept_ + model.coef_ * x_new

plt.plot(x_new, y_pred)

plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('MSE = ', mean_squared_error(y_test, model.predict(x_test)))
print('R^2 = ', r2_score(y_test, model.predict(x_test)))
```
